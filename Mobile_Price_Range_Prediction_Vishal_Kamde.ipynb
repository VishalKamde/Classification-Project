**pixel_width**

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
sns.kdeplot(data=df, x='px_width', hue='price_range', ax=axs[0])
sns.boxplot(data=df, x='price_range', y='px_width', ax=axs[1])
plt.show()
```

The pixel width does not exhibit a continuous increase when transitioning from low cost to very high cost. Mobiles with both "Medium cost" and "High cost" have nearly equal pixel width, indicating that it could be a significant determining factor in the price range.

**pixel Height**

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
sns.kdeplot(data=df, x='px_height', hue='price_range', ax=axs[0])
sns.boxplot(data=df, x='price_range', y='px_height', ax=axs[1])
plt.show()
```
